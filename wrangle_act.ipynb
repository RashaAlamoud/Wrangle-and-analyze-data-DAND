```python
# Make copies of dataframes to clean 
clean_twitter_archive = twitter_archive.copy()
clean_image_predictions = image_predictions.copy()
df_clean_tweet_info = df_tweet_info.copy()
```

#### Quality issue1:Tweet_id is an integer.

```python
# Change 'tweet_id' to a string
clean_twitter_archive['tweet_id'] = twitter_archive['tweet_id'].astype(str)
clean_twitter_archive.info()
```

#### Quality issue 2:Dog names starting with a lowercase letter 

```python
# Find all names that start with a lowercase letter
lowercase_names = []
for row in twitter_archive['name']:
    if row[0].islower() and row not in lowercase_names:
        lowercase_names.append(row)
print(lowercase_names)
```

```python
# Replace all names that start with a lowercase letter with a NaN
clean_twitter_archive=twitter_archive['name'].replace(lowercase_names, 
                        np.nan,
                       inplace = True)

twitter_archive['name'].value_counts()
```

#### Quality issue 3:Timestamp is a string

```python
#Convert timestamp to datetime object
twitter_archive['timestamp'] = pd.to_datetime(twitter_archive['timestamp'])
twitter_archive.info()
```

#### Quality issue 4: Remove retweets

```python
# Keep columns where 'retweeted_status_id' is NaN
twitter = twitter_archive[np.isnan(twitter_archive.retweeted_status_id)]
# Remove 'retweeted_status_id', 'retweeted_status_user_id' and 
# 'retweeted_status_timestamp' columns
twitter = twitter_archive.drop(['retweeted_status_id', 
                        'retweeted_status_user_id', 
                        'retweeted_status_timestamp'], 
                       axis=1)
twitter_archive.info()
```

#### Quality issue 5: Incorrect values in rating numerators 

```python
# Calulate the value of 'rating'
twitter_archive['rating'] = twitter_archive['rating_numerator'] / twitter_archive['rating_denominator']
twitter_archive.head()
```

####  Quality issue 6:Name being "O" instead of "O'Malley" 

```python
# Replace the name 'O' with "O'Malley" 
twitter_archive['name'].replace('O', 
                        "O'Malley",
                       inplace = True)

twitter_archive['name'].value_counts()
```

```python
# merge the doggo, floofer, pupper and puppo columns to a stage column
twitter_archive['stage'] = twitter_archive[['doggo', 'floofer', 'pupper', 'puppo']].max(axis=1)

# convert the datatype from string to categorical
twitter_archive.stage = twitter_archive.stage.astype('category')

# drop the doggo, floofer, pupper and puppo columns
twitter_archive.drop(['doggo', 'floofer', 'pupper', 'puppo'], axis=1, inplace=True)
```

#### Tidiness issue 2:Three data frames twitter_archive, image_predictions, and df_tweet_info should be one (combined table)

```python
# Merge the three dataframes on 'tweet_id'
dfs = [clean_twitter_archive, clean_image_predictions, df_clean_tweet_info]
twitter = reduce(lambda left,right: pd.merge(left,right,on='tweet_id'), dfs)
```
